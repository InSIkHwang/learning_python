### Pandas 데이터 관리(https://pandas.pydata.org/pandas-docs/stable/index.html)


```python
# 필요한 모듈들을 연결한다.

import pandas as pd
import numpy as np
```

```python
# LOCAL_PEOPLE_GU_2023.csv 파일 읽어오기

df1 = pd.read_csv("./data/LOCAL_PEOPLE_GU_2023.csv", encoding="cp949")
df1.head()   # head(숫자) : df에서 입력한 숫자 만큼 위쪽부터 데이터 출력, 기본값 5
```

```python
# df.info() 데이터 정보 확인
df1.info()
```

```python
# columns 이름 확인
df1.columns
```

```python
# columns 이름 변경하기
df1.columns = ['기준일', '시간대', '자치구', '총생활인구수', '남자0-9', '남자10-14',
       '남자15-19', '남자20-24', '남자25-29','남자30-34', '남자35-39', '남자40-44',
       '남자45-49', '남자50-54', '남자55-59','남자60-64', '남자65-69', '남자70이상', 
       '여자0-9','여자10-14', '여자15-19', '여자20-24','여자25-29', '여자30-34', '여자35-39',
       '여자40-44', '여자45-49', '여자50-54','여자55-59', '여자60-64', '여자65-69','여자70이상']

df1.tail(3) # tail(숫자) : df에서 입력한 숫자 만큼 아래쪽부터 데이터 출력, 기본값 5
```

```python
# 데이터 삭제: drop()
# drop()을 이용한 열 삭제: df.drop(columns="삭제필드") / df.drop(columns=["삭제필드1", "삭제필드2"])
df1 = df1.drop(columns=["자치구"]) # 삭제된 결과 즉시 적용: inplace=True
df1.head(3)
```

```python
# 계산에 의한 필드 추가(계산필드): df.assign(추가필드명 = 계산식) 
df1.assign(_10대이하 = df1["남자0-9"] + df1["여자0-9"]).head()
```

```python
# 계산에 의한 필드 추가(계산필드): 리스트 형태로 추가
tmp = []
for i in df1.index:
  tmp.append(np.sum(df1.iloc[i, 4:18]))

df1["남자생활인구수"] = tmp
df1.head(3)
```

```python
tmp = []
for i in df1.index:
  tmp.append(np.sum(df1.iloc[i, 18:-1]))

df1["여자생활인구수"] = tmp
df1.head(3)
```

```python
# 필드단위 계산: numpy, df.agg(변수=("계산필드", "계산함수명"), ...)
npVal=np.mean(df1["남자생활인구수"].values)   # 단일 필드 계산
print(npVal)

agg1=df1.agg(m_mean=("남자생활인구수", "mean"))
# print(agg1)
display(agg1)   # 형식을 가지는 데이터 출력

agg2=df1.agg(m_mean=("남자생활인구수", "mean"),
             f_mean=("여자생활인구수", "mean"))
display(agg2)

```

```python
# 집단별 요약(그룹별 계산-단일):
# => df.groupby(["필드명"])[["계산필드", ...]].mean() / sum(), max(), min()
# => df.groupby(["필드명"]).agg(변수=("계산필드", "계산함수명"), ...)
group1 = df1.groupby(["기준일"])[["남자생활인구수","여자생활인구수"]].mean()
display(group1.head())

group2 = df1.groupby(["기준일"]).agg(m_mean=("남자생활인구수", "mean"),
                                    f_mean=("여자생활인구수", "mean"))
display(group2.head())
```

자주 사용하는 통계량<br>




```python
# 집단별 요약(그룹별 계산-다중):
# => df.groupby(["필드명", ...])[["계산필드", ...]].mean() / sum(), max(), min()
# => df.groupby(["필드명", ...]).agg(변수=("계산필드", "계산함수명"), ...)

group1 = df1.groupby(["기준일","시간대"])[["남자생활인구수","여자생활인구수"]].mean()
display(group1.head())

group2 = df1.groupby(["기준일","시간대"]).agg(남자평균=("남자생활인구수", "mean"),
                                             남자합계=("남자생활인구수", "sum"))
display(group2.head())
```

```python
# 특정 조건에 맞는 값만 추출해 통계량 계산
# \ : 코드가 긴 경유 줄바꾸기시 사용

query1=df1.query("시간대 == 7").groupby(["기준일"])[["남자생활인구수","여자생활인구수"]].mean()
display(query1)

query2 = df1.query("시간대 == 7").groupby(["기준일","시간대"])\
  .agg(남자평균=("남자생활인구수", "mean"),
       남자합계=("남자생활인구수", "sum"))
display(query2)
```

```python
# 데이터 정렬: 
# 값 정렬: df.sort_values(["필드명", ...], ascending=False) => ascending => False(내림차순)/True(오름차순, 기본값)
# index 정렬: df.sort_index(ascending=False)

df2 = df1[["기준일", "시간대", "총생활인구수", "남자생활인구수",	"여자생활인구수"]]
display(df2.head())

df2=df2.sort_values(["총생활인구수"])  # 총생활인구수를 기준으로 오름차순 정렬후 df2에 저장
display(df2.head())

df2=df2.sort_values(["총생활인구수"], ascending=False)  # 총생활인구수를 기준으로 내림차순 정렬후 df2에 저장
display(df2.head())

df2=df2.sort_index()
display(df2.head())
```
