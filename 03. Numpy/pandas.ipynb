### Pandas

pandas 자료구조

- DataFrame: 2차원 배열 구조

```python
import pandas as pd

tmp = [["KIM",92,68,89],["LEE",78,98,66],["PARK",95,74,76]]
df1 = pd.DataFrame(tmp, columns=["이름","국어", "영어","수학"])
df1
```

pandas 파일 읽기/쓰기

- pd.read_csv(FilePath, sep, header, names, index_col, skiprows, nrows, encoding)
- FilePath : 파일경로 , sep : 구분자 , header(컬럼명) : 없을 경우 none
- names : header가 없을시 컬럼명 입력 가능 , skiprows : 파일에서 행을 건너뛰고 불러옴
- index_col : 컬럼을 index로 사용 , nrows : 입력한 개수만큼의 데이터만 읽음
- encoding : 인코딩 타입 입력, 데이터가 한글이면 'CP949'

### 데이터 조회


열 기준 조회
- 1개 열 조회 : df["필드명"]
- 2개 이상 열 조회 : df[["필드명1", "필드명2", ...]]

```python
df1["상호"]
```

```python
df2 = df1[['상표','상호', '주소',  '전화번호', '셀프여부','휘발유', '경유']]
df2
```

행 기준 출력
- index 이용 
  - index => 행 ID
- 위치 값 이용
  - 데이터의 인덱싱 위치 값

```python
df2.loc[3:7]
```

```python
df2.loc[3:7, ["상표", "휘발유", "경유"]]
```

```python
df2 = df2.sort_values("휘발유")
df2
```

```python
df2.loc[20:21, ["상호", "휘발유"]]
```

```python
df2.iloc[0:3, 4:7]
```
