# Python 데이터 자료형

### 변수
* 영어 + 숫자 등을 사용하여 생성  
* 대소문자 구분  
* 문자부터 시작해야 하며 숫자가 먼저 나오면 안 됨  
* 특수기호는 Underscore(_) 만 사용 가능  
* 예약어( if , for , while , def , end , or 등) 는 사용 불가능  

```python
name = "KIM"
age = 55

print(f"{name}님의 나이는 {age}입니다.")
print("%s님의 나이는 %.2f입니다."%(name, age))
```

### 숫자형
* 정수 (integer) : 양의정수, 음의정수, 숫자 0 ex) 123, -456, 0
* 실수 (float) : 소수점이 포함된 숫자 ex) 123.1, -323.3
* 8진수 (octal) : 0o11, 0o323
* 16진수 (hexadecimal) : 0x2F, 0x3D

```python
# type(): 자료형 반환
int1=123
float1=1.23
float2=1.

print(type(int1), type(float1), type(float2))
```

### 문자열

* 문자나 단어등으로 이루어진 문자들의 집합
* 문자열은 큰따옴표 ( “ ” ) 또는 작은 따옴표 ( ‘ ’ ) 로 둘러싸여 생성

```python
str1 = "123"

print(type(str1))
```

```python
str2 = "MultiCampus Front-end"

print(str2[6])
print(str2[-1])
print(str2[0:12])
print(str2[-9:])

```

### 리스트와 배열
리스트 : 데이터들의 모음 , 대괄호 사용 , 각 데이터는 ( , ) 로 구분

```python
lst1=[1,2,3,4,5]
lst2=["Lee","Hong","Park","Kim","Seo"]
lst3 = [10,20,"multi","campus",[30,40,50]]

print(type(lst1), type(lst2),type(lst3))

```

```python
lst3[2] ="Multi"
print(lst3)

print(lst3[-1])
print(lst3[-1][0])
```

```python
lst2.append("HWANG")
print(lst2)

del lst2[5]
print(lst2)

lst2.append("HWANG")
print(lst2)

lst2.remove("HWANG")
print(lst2)
```

```python
lst2.sort()
print(lst2)

lst2.reverse()
print(lst2)
```

```python
import numpy as np 

arr1=np.array(lst1)
arr2=np.array(lst2)

print(type(arr1), arr1, lst1)
print(type(arr2), arr2, lst2)
```
